```python
import pandas as pd
```

```python
data = {
    "resume": [
        "Python machine learning data science projects internship",
        "No skills basic resume fresher",
        "Java developer with internship and web development experience",
        "Fresher with no coding knowledge",
        "AI ML deep learning and NLP projects",
        "Only theoretical knowledge no practical experience",
        "Full stack web developer React Node projects",
        "No technical skills only degree"
    ],
    "status": ["Selected", "Rejected", "Selected", "Rejected", "Selected", "Rejected", "Selected", "Rejected"]
}

df = pd.DataFrame(data)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df["resume"], df["status"], test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train_vec, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test_vec)
print("Model Accuracy:", accuracy_score(y_test, y_pred))
```

```python
user_input = input("Enter resume text: ")
user_vec = vectorizer.transform([user_input])

result = model.predict(user_vec)
print("Prediction:", result[0])
```

```python
import matplotlib.pyplot as plt

# Count Selected vs Rejected
df["status"].value_counts().plot(kind='bar')

plt.title("Resume Selection Distribution")
plt.xlabel("Status")
plt.ylabel("Count")

plt.show()
```

```python
keywords = ["python", "machine learning", "data science", "ai", "ml", "java", "react", "node"]

matched = []

for word in keywords:
    if word in user_input.lower():
        matched.append(word)

if len(matched) > 0:
    print("Reason: Contains keywords like", ", ".join(matched))
else:
    print("Reason: Lacks strong technical keywords")
```
